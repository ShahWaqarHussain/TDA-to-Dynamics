# V-Construction of gray_images

```python
import matplotlib.pyplot as plt
from persim import plot_diagrams

# Assume `img2d` is your grayscale image, and `cr.computePH` computes the persistent homology
img2d = gray_img0
pd = cr.computePH(img2d)
pds = [pd[pd[:, 0] == i] for i in range(2)]  # Separate 0- and 1-dimensional persistence

# Print Betti numbers for each dimension
print("Betti numbers: ", [len(pds[i]) for i in range(len(pds))])

# Font size for labels and tick values
label_font_size = 20
tick_font_size = 18

# Plot 0-dimensional and 1-dimensional persistence diagrams separately
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 6))

# 0-D Persistence Diagram
plot_diagrams([pds[0][:, 1:3]], ax=ax1, lifetime=True, legend=False)
#ax1.set_title('0-D Persistence Diagram', fontsize=label_font_size)
ax1.set_xlabel("Birth", fontsize=label_font_size)
ax1.set_ylabel("Death", fontsize=label_font_size)
ax1.tick_params(axis='both', which='major', labelsize=tick_font_size)
ax1.collections[0].set_color("blue")  # Set color for 0-D diagram

# 1-D Persistence Diagram
plot_diagrams([pds[1][:, 1:3]], ax=ax2, lifetime=True, legend=False)
#ax2.set_title('1-D Persistence Diagram', fontsize=label_font_size)
ax2.set_xlabel("Birth", fontsize=label_font_size)
ax2.set_ylabel("Death", fontsize=label_font_size)
ax2.tick_params(axis='both', which='major', labelsize=tick_font_size)
ax2.collections[0].set_color("red")  # Set color for 1-D diagram

plt.tight_layout()
plt.show()
```

# T-Construction of gray_ images

```python
# compute PH for the T-construction of the original image (pixel value filtration)
img2d=gray_img0
pdt = tcr.computePH(img2d)
pdst = [pdt[pdt[:,0] == i] for i in range(2)]
print("Betti numbers: ",[len(pdst[i]) for i in range(len(pdst))])

persim.plot_diagrams([p[:,1:3] for p in pdst],title='T-construction')
```

```python
import matplotlib.pyplot as plt
from persim import plot_diagrams

# Assume `img2d` is your grayscale image, and `cr.computePH` computes the persistent homology
img2d = gray_img0
pdt = tcr.computePH(img2d)
pdst = [pdt[pdt[:,0] == i] for i in range(2)]
print("Betti numbers: ",[len(pdst[i]) for i in range(len(pdst))])

# Font size for labels and tick values
label_font_size = 20
tick_font_size = 18

# Plot 0-dimensional and 1-dimensional persistence diagrams separately
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 6))

# 0-D Persistence Diagram
plot_diagrams([pdst[0][:, 1:3]], ax=ax1, lifetime=True, legend=False)
#ax1.set_title('0-D Persistence Diagram', fontsize=label_font_size)
ax1.set_xlabel("Birth", fontsize=label_font_size)
ax1.set_ylabel("Death", fontsize=label_font_size)
ax1.tick_params(axis='both', which='major', labelsize=tick_font_size)
ax1.collections[0].set_color("blue")  # Set color for 0-D diagram

# 1-D Persistence Diagram
plot_diagrams([pdst[1][:, 1:3]], ax=ax2, lifetime=True, legend=False)
#ax2.set_title('1-D Persistence Diagram', fontsize=label_font_size)
ax2.set_xlabel("Birth", fontsize=label_font_size)
ax2.set_ylabel("Death", fontsize=label_font_size)
ax2.tick_params(axis='both', which='major', labelsize=tick_font_size)
ax2.collections[0].set_color("red")  # Set color for 1-D diagram

plt.tight_layout()
plt.show()
```
