```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import ripser as ripser
from persim import plot_diagrams
```

```python
image = cv2.imread('C:\\Users\\HP\\OneDrive\\project with Rouhi\\dataset\\butterfly.png') # download the image from the dataset and give you own path of the image
plt.imshow(image)
plt.axis('off')  # Hide axes
plt.show()
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
plt.axis('off')  # Hide axes
plt.show()
```

```python
edges = cv2.Canny(gray, 100, 200)

# Get coordinates of edge points
points = np.column_stack(np.where(edges > 0))

# Plot the points
plt.scatter(points[:, 1], points[:, 0], s=1)
plt.gca().invert_yaxis()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D


# Extracting the x, y, and z coordinates
x = points[:, 0]
y = points[:, 1]
z = np.zeros_like(x)  # Assuming z-coordinates are all zeros or another set of values

# Creating the 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z)

# Adding labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Scatter Plot of Points')

# Displaying the plot
plt.show()
```

```python

diagrams = ripser(points,n_perm=1200,thresh=156)['dgms']

# Plot the persistence diagram
plot_diagrams(diagrams)
```
