```python
# we use the ripser library for the compuation of homoloy
res = ripser(merge)


# persistence diagrams are stored in the dictionary in the output of the ripser7
pd0 = res['dgms'][0]
pd1= res['dgms'][1]
```

```python
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
plot_diagrams(res['dgms'], lifetime=False)
```

```python
import matplotlib.pyplot as plt

# Assuming res['dgms'][0] and res['dgms'][1] are lists of tuples (birth, death)

# Extracting birth and death coordinates for 0-dimensional persistence diagram
birth_death_0 = res['dgms'][0]
births_0, deaths_0 = zip(*birth_death_0)  # Separate birth and death values
max_val_0 = max(max(births_0), max(deaths_0))  # Maximum value for diagonal line

# Plot for 0-dimensional persistence diagram
plt.figure(figsize=(8, 6))
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
plt.ticklabel_format(axis='y', style='sci', scilimits=(0,0))  # Scientific notation for y-axis
plt.scatter(births_0, deaths_0, color='blue')
plt.plot([0, max_val_0], [0, max_val_0], 'k--')  # Diagonal line
plt.title("0-Dimensional Persistence Diagram", fontsize=24)
plt.legend()
plt.show()

# Extracting birth and death coordinates for 1-dimensional persistence diagram
birth_death_1 = res['dgms'][1]
births_1, deaths_1 = zip(*birth_death_1)  # Separate birth and death values
max_val_1 = max(max(births_1), max(deaths_1))  # Maximum value for diagonal line

# Plot for 1-dimensional persistence diagram
plt.figure(figsize=(8, 6))
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
plt.scatter(births_1, deaths_1, color='green')
plt.plot([0, max_val_1], [0, max_val_1], 'k--')  # Diagonal line
plt.title("1-Dimensional Persistence Diagram", fontsize=24)
plt.legend()
plt.show()
```
